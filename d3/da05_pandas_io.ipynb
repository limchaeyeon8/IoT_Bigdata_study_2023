```python
import pandas as pd
import numpy as np
```

```python
# 로컬파일 읽기 // 속성(option) 중 encoding은 필수 (euc-kr | utf-8) , header속성 기본 = 0
df_chb_foodbank = pd.read_csv('../raws/raw/충청북도_푸드뱅크현황.csv', encoding='utf-8', header=0)
df_chb_foodbank
```

```python
# info() 결측치도 확인 / 컬럼갯수 / 데이터 갯수
df_chb_foodbank.info()
```

```python
# dtype이 int, float이 없으면 통계치 못 냄
df_chb_foodbank.describe()
```

```python
# 데이터를 저장
df_chb_foodbank.to_csv('./data/충북푸드뱅크현황.csv', encoding='utf-8', index=False )
```
